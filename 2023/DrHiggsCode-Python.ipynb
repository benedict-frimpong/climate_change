```python
# Extract minimums and maximums

def minmax(data):
    i_anchors = []
    val_anchors = []
    for i in range(2, len(data)): # if statement started at 3, so included that start up here instead
        # needs a continuity check
        # needs a normalization because observation will likely miss actual min/max
        if (data[i] - data[i-1]) > 0:
            if (data[i-1] - data[i-2]) < 0:
                # minimum located
                i_anchors.append(i-1)
                val_anchors.append(1)
            elif (data[i-1] - data[i-2]) == 0:
                # zero slope located
                if (i >= 3) and (data[i-2] - data[i-3]) < 0:
                    #minimum located
                    i_anchors.append(i-1)
                    val_anchors.append(1)
            else:
                continue
        elif (data[i] - data[i-1]) < 0:
            if (data[i-1] - data[i-2]) > 0:
                # maximum located
                i_anchors.append(i-1)
                val_anchors.append(2)
            elif data[i-1] - data[i-2] == 0:
                # zero slope located
                if i >= 4 and (data[i-2] - data[i-3]) > 0:
                    #maximum located
                    i_anchors.append(i-2)
                    val_anchors.append(2)
            else: continue
        else: continue
    
    anchors = pd.DataFrame({'index':i_anchors, 'value':val_anchors})
    return anchors
```

```python
def avg(df):
    arr = []
    for i in range(len(df.columns)):
        val = df[i].sum()
        arr.append(val)
    for j in range(len(arr)):
        arr[j] = float(arr[j])/len(df)
    return arr

masterkey_test = avg(masterset_test)
display(masterkey_test)
```

```python
# Calculate masterset error

def master_error(df, key):
    df_array = []
    for i in range(len(df)):
        # cycle through each set in the masterset
        arr = []
        for m in range(len(df.columns)):
            val = df[m][i] - key[m]
            arr.append(val)
        df_array.append(arr)
    
    df_array = pd.DataFrame(df_array)
    return df_array

mastererror_test = master_error(masterset_test, masterkey_test)
display(mastererror_test)
```
